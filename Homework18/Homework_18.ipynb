```python
import pandas as pd
df = pd.read_csv('tackoverflow_qa.csv')
df.head()
```

```python
#1. Find all questions that were created before 2014
df[df['creationdate'] < '2014-01-01 00:00:00']
```

```python
#2. Find all questions with a score more than 50
df[df['score'] > 50]
```

```python
#3. Find all questions with a score between 50 and 100
df[(df['score'] > 50) & (df['score'] < 100)]
```

```python

#4. Find all questions answered by Scott Boston
df[df['ans_name'] == 'Scott Boston']
```

```python
#5. Find all questions answered by the following 5 users
df[df['ans_name'].isin(['Scott Boston', 'Jason Strimpel', 'David Underhill', 'DarkAnt', 'tatwright'])]
```

```python
#6. Find all questions that were created between March, 2014 and October 2014 that were answered by Unutbu and have score less than 5.
df[(df['creationdate'] > '2014-03-01 00:00:00') & (df['creationdate'] < '2014-10-31 23:59:59') & (df['ans_name'] == 'Unutbu') & (df['score'] < 5)]
```

```python
#7. Find all questions that have score between 5 and 10 or have a view count of greater than 10,000
df[(df['score'] > 5) & (df['score'] < 10) | (df['viewcount'] > 10000)]
```

```python
#8. Find all questions that are not answered by Scott Boston
df[df['ans_name'] != 'Scott Boston']
```

```python
titanic_df = pd.read_csv("titanic.csv")
titanic_df.head()
```

```python
# 1. Select Female Passengers in Class 1 with Ages between 20 and 30
female_class1_20_30 = titanic_df[(titanic_df['Sex'] == 'female') & 
                                 (titanic_df['Pclass'] == 1) & 
                                 (titanic_df['Age'] >= 20) & 
                                 (titanic_df['Age'] <= 30)]

print(female_class1_20_30)
```

```python
# 2. Filter Passengers Who Paid More than $100
paid_more_than_100 = titanic_df[titanic_df['Fare'] > 100]
print(paid_more_than_100)
```

```python
# 3. Select Passengers Who Survived and Were Alone
survived_alone = titanic_df[(titanic_df['Survived'] == 1) & 
                            (titanic_df['SibSp'] == 0) & 
                            (titanic_df['Parch'] == 0)]
print(survived_alone)
```

```python
# 4. Filter Passengers Embarked from 'C' and Paid More Than $50
embarked_c_paid_50 = titanic_df[(titanic_df['Embarked'] == 'C') & 
                                (titanic_df['Fare'] > 50)]
print(embarked_c_paid_50)
```

```python
# 5. Select Passengers with Siblings or Spouses and Parents or Children
siblings_spouses_and_parents_children = titanic_df[(titanic_df['SibSp'] > 0) & 
                                                   (titanic_df['Parch'] > 0)]

print(siblings_spouses_and_parents_children)
```

```python
# 6. Filter Passengers Aged 15 or Younger Who Didn't Survive
age_15_or_younger_didnt_survive = titanic_df[(titanic_df['Age'] <= 15) & 
                                             (titanic_df['Survived'] == 0)]

print(age_15_or_younger_didnt_survive)
```

```python
# 7. Select Passengers with Cabins and Fare Greater Than $200
cabins_and_fare_200 = titanic_df[titanic_df['Cabin'].notna() & 
                                 (titanic_df['Fare'] > 200)]

print(cabins_and_fare_200)
```

```python
# 8. Filter Passengers with Odd-Numbered Passenger IDs
odd_passenger_ids = titanic_df[titanic_df['PassengerId'] % 2 != 0]

print(odd_passenger_ids)
```

```python
# 9. Select Passengers with Unique Ticket Numbers
unique_tickets = titanic_df[titanic_df['Ticket'].duplicated(keep=False) == False]
print(unique_tickets)
```

```python
# 10. Filter Passengers with 'Miss' in Their Name and Were in Class 1
miss_in_name_class1 = titanic_df[(titanic_df['Name'].str.contains('Miss')) & 
                                 (titanic_df['Pclass'] == 1)]
print(miss_in_name_class1)
```
